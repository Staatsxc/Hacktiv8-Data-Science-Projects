### LOADING DATA 

```python
import pandas as pd
import numpy as np
df = pd.read_csv("h8dsft_P0GC2_Set_1_Alexander_GT.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.tail()
```

### Membuat Variable

```python
start_stations = df.loc[:,'start_station_name'].unique()
end_stations = df.loc[:,'end_station_name'].unique()
exclusive_stations = np.setxor1d(start_stations,end_stations)
```

```python
exclusive_stations.shape
```

### Memasukkan data kedalam matrix

```python
df_matrix = pd.crosstab( df.loc[:,'start_station_name'],df.loc[:,'end_station_name'])
df_matrix.shape
```

## Membuat df copy

```python
df_copy = df.copy()
```

```python
df_copy = df_copy[~(df_copy['start_station_name'].isin(exclusive_stations))&
                ~(df_copy['end_station_name'].isin(exclusive_stations))]
```

```python
df_copy
```

## Membuat crosstab dari df copy

```python
df_matrix_square = pd.crosstab(df_copy.loc[:,'start_station_name'],df_copy.loc[:,'end_station_name'])
df_matrix_square
```

```python
total_trips = df_matrix_square.sum(axis=1)
```

## Membuat weighted matrix dengan membagi data dengan rows nya

```python
weighted_matrix_df = df_matrix_square.div(total_trips, axis = 'rows')
```

```python
start = list(df_matrix_square.index)
type(start)
```

```python
weighted_matrix_df
```

## Membuat matrix menjadi array

```python
array = np.array(weighted_matrix_df)
```

## Mencari nilai eigen value dan vector

```python
eigen_value, eigen_vector = np.linalg.eig(array)
np.argmax(eigen_value)
```

```python
eigen_value.max()
```

## Memasukkan hasil yang sudah diubah menjadi absolute

```python
PR_Scores = np.abs(list(eigen_vector[26]))
PR_Scores
```

## Membuat df pagerank

```python
pagerank_df = pd.DataFrame({'Start_stations': start,
                   'PR_Scores': PR_Scores})
pagerank_df
```

```python
# Mencari lokasi dimana pagerank terbesar berada
np.argmax(np.abs(eigen_vector[26]))
```

Eigen decomposition pada kasus ini kita gunakan untuk mencari stasiun mana dengan traffic yang paling tinggi. eigen decomposition berguna karena eigen decomposition dapat mencari eigen value dan eigen vector nya. eigen value merupakan value dimana sebuah garis di perbesar, maka kita cari yang value nya paling tinggi dan score dari pagerank nya adalah eigen vector. eigen vector berguna untuk mengurangi operasi aljabar linear dan dapat memisahkan masalah menjadi lebih simpel.

```python
# Membuat Df baru yang region id nya 3
df_coor_stations = df['region_id'] == 3
```

## Load library yang akan dipakai dan menyiapkan dataframe

```python
from scipy.integrate import trapz
from scipy import integrate
df_coor_stations = df[(df['region_id'] == 3)]
df_coor_stations = df_coor_stations.drop(['trip_id','start_date','end_station_name',], axis = 1)
df_coor_stations = df_coor_stations.sort_values(by = 'start_station_longitude')
```

## Membuat plot data

```python
df_coor_stations.plot(x = 'start_station_longitude', y = 'start_station_latitude')
```

## Membuat sumbu x dan y lalu menghitung luas dengan integral trapezoid

```python
sumbu_x = df_coor_stations['start_station_longitude']
sumbu_y = df_coor_stations['start_station_latitude']
```

```python
luas = integrate.trapz(sumbu_y,sumbu_x)
print('Luas stasiun di region 3 adalah:',luas)
```

1. Terdapat 2 jenis integral yaitu definite dan indefinite. Jenis integral apa yang kamu gunakan untuk kasus ini? Berikan alasanmu!

integral yang digunakan dalam kasus ini adalah integral definite karena integral definite dapat mengembalikan hasil berupa scalar sedangkan integral indefinite tidak dapat mengembalikan hasil berbentuk scalar. berbeda seperti definite integral, indefinite integral hanya return integral fungsinya sedangkan definite integral dapat mengembalikan hasil scalar seperti area.

2. Terdapat dua metode dalam perhitungan integral yaitu simbolik dan numerik. Mana yang kamu gunakan untuk kasus ini dan mengapa?

perhitungan integral yang digunakan merupakan perhitungan integral numerik karena penghitungan definite integral hanya bisa menggunakan perhitungan integral numerik. perhitungan integral simbolik tidak dapat menghasilkan skalar seperti luas.
