```python
def plot_age(df, col):
    fig, axes = plt.subplots(ncols = 2, figsize = (15, 5))
    
    # histogram
    sns.histplot(df[col],ax = axes[0], kde=True,line_kws={'linewidth': 3},color = 'k', bins = 20)
    axes[0].set_title(f"Histogram '{col}'")
    axes[0].axvline(df[col].mean(), color = 'red', linestyle = 'dashed', label = 'mean')
    axes[0].axvline(df[col].median(), color = 'green', linestyle = 'dashed', label = 'median')
    axes[0].legend()
    axes[0].xaxis.set_major_locator(MultipleLocator(5))
    
    # boxplot
    sns.boxplot(y=df[col], ax =  axes[1])
    axes[1].set_title(f"Boxplot '{col}'")
    
    plt.show()
    
    # skewness
    print(df[col].name + ' Kurtosis: ' + str(df[col].kurt()))
    print(df[col].name + ' Skewness: ' + str(df[col].skew()))
    if -0.5 <= df[col].skew() <= 0.5:
        print("Columns '{}' normal distribution".format(col))
    elif df[col].skew() > 0.5:
        print("Columns '{}' right skewed".format(col))
    elif df[col].skew() < -0.5:
        print("Columns '{}' left skewed".format(col))
```

```python
plot_age(df,'age')
```

## Outlier Handling

```python
mean = df.age.mean()
standard_dev = df.age.std()
print(mean)
print(standard_dev)
```

```python
upper_limit = mean + 3*standard_dev
upper_limit
```

```python
lower_limit = mean - 3*standard_dev
lower_limit
```

```python
outliers = df [(df.age < lower_limit) | (df.age > upper_limit)]
outliers.head()
```

This is the outliers data of age that is unlikely to happen, Data scientist tend to remove data that is unlikely to happen to give a better analytics insight and only focus on the majority of Data

```python
# Making a clean dataset without outliers
df_clean = df [(df.age > lower_limit) & (df.age < upper_limit)]
df_clean.reset_index(drop = True, inplace = True)
```

```python
# Giving value of 1 for each crash case
df_clean['Case_Num'] = 1
```

```python
df_clean.head()
df_clean.to_csv("df_clean.csv")
```

```python
plot_age(df_clean,'age')
```

## Crash Case based on Days

## Inferential Statistics Analysis

For this Analysis, Data Scientist will do hypothesis testing of correlation between gender of the driver and the Fatal injury caused, the data scientist choose fatal injury because fatal injury case is the most dangerous injury out of all and it is also the most happened injury compared to any other injury

```python
# Making Dataset of only driver
df_driver = df_clean[df_clean['person_type_name'] == "Driver of a Motor Vehicle In-Transport"]
df_driver.head()
```

```python
chi_df= df_clean[(
    df_clean['injury_severity_name'] == 'Fatal Injury (K)')|
    (df_clean['injury_severity_name'] == 'No Apparent Injury (O)')]
```

```python
chi_df = chi_df[chi_df.sex != 'Not Reported']
```

```python
contingency_table = pd.crosstab(chi_df["sex"],chi_df["injury_severity_name"])
contingency_table
```

```python
contingency_table.to_csv('contingency.csv')
```

```python
stat, p, dof, expected = stats.chi2_contingency(contingency_table)

print('p=%.3f' % (p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent')
```

### Result of Statistical Test

The probability value is more than 0.05 or alpha, it means that it is very possible that the fact male cause more fatal injury than female is only a coincidence. there is no statistical significance of relation between male and female drive to the injury caused.
